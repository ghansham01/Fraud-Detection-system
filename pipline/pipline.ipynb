```python
counts = df["Class"].value_counts()
fraud_pct = counts[1] / len(df) * 100

print(f"Legit txns: {counts[0]:,}")
print(f"Fraud txns: {counts[1]:,}")
print(f"Fraud rate: {fraud_pct:.4f}%  ← severe imbalance")
```

```python
missing = df.isnull().sum().sum()
print(f"Missing vals : {missing}")
```

```python

print("\nClass distribution:")
print(counts)
print(f"Fraud ratio: {counts[1] / counts.sum():.6f}")
```

```python
df = df.dropna(subset=["Class"])

before = len(df)

numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()


for col in numeric_cols:
    if df[col].isnull().any():
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
        print(f"Filled '{col}' nulls with median={median_val:.4f}")

after = len(df)
print(f"Rows before: {before:,}")
print(f"Rows after: {after:,}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df["Amount"], bins=50, ax=axes[0], kde=True)
axes[0].set_title("Amount distribution")
sns.histplot(df["Time"], bins=50, ax=axes[1], kde=True)
axes[1].set_title("Time distribution")
plt.tight_layout()
plt.show()
```

```python
corr = df.corr()["Class"].sort_values()
print("\nTop correlations with Class:")
print(corr.tail(10))

plt.figure(figsize=(10, 6))
sns.heatmap(df.loc[:, ["Time", "Amount"] + [c for c in df.columns if c.startswith("V")]].corr(), cmap="coolwarm", center=0)
plt.title("Feature correlation matrix")
plt.show()
```
